```python
# Importar librerias pandas, numpy, matpltlib y seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/spamella04/InteligenciadeNegocios2024/refs/heads/IvaChavarria/ClasePractica1/public-health-statistics-selected-public-health-indicators-by-chicago-community-area-1.csv'
df=pd.read_csv(url)
df.head()
```

```python
df.fillna(0, inplace=True)
```

```python
df_numerico = df.select_dtypes(include=['float64', 'int64'])
```

```python
media = df_numerico.mean()
print(media)
```

```python
media.plot(kind='bar', figsize=(10, 6))
plt.title('Media de las columnas')
plt.ylabel('Media')
plt.xticks(rotation=90)
plt.show()
```

```python
media_sin_per_capita = media.drop('Per Capita Income')
media_sin_per_capita.plot(kind='bar', figsize=(10, 6))
plt.title('Media de las columnas sin Per Capita Income')
plt.ylabel('Media')
plt.xticks(rotation=90)
plt.show()
```

```python
mediana = df_numerico.median()
print(mediana)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_numerico)
plt.title('DistribuciÃ³n de variables numÃ©ricas de Salud PÃºblica')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_numerico.drop('Per Capita Income', axis=1))
plt.title('DistribuciÃ³n de Variables NumÃ©ricas de Salud PÃºblica (sin Per Capita Income)')
plt.xticks(rotation=90)
plt.show()
```

```python
desviacion_estandard = df_numerico.std()
print(desviacion_estandard)
```

```python
desviacion_estandard.plot(kind='bar', figsize=(10, 6))
plt.title('DesviaciÃ³n EstÃ¡ndar de Salud PÃºblica')
plt.ylabel('DesviaciÃ³n EstÃ¡ndar')
plt.xticks(rotation=90)
plt.show()
```

```python
desviacion_estandar_sin_per_capita = desviacion_estandard.drop('Per Capita Income')
desviacion_estandar_sin_per_capita.plot(kind='bar', figsize=(10, 6))
plt.title('DesviaciÃ³n EstÃ¡ndar de Variables NumÃ©ricas (sin Per Capita Income)')
plt.ylabel('DesviaciÃ³n EstÃ¡ndar')
plt.xticks(rotation=90)
plt.show()
```

```python
minimo = df_numerico.min()
print(minimo)
```

```python
minimo.plot(kind='bar', figsize=(10, 6))
plt.title('Valor MÃ­nimo de Variables NumÃ©ricas de Salud PÃºblica')
plt.ylabel('Valor MÃ­nimo')
plt.xticks(rotation=90)
plt.show()
```

```python
minimo_sin_per_capita = minimo.drop('Per Capita Income')
minimo_sin_per_capita.plot(kind='bar', figsize=(10, 6))
plt.title('Valor MÃ­nimo de Variables NumÃ©ricas (sin Per Capita Income)')
plt.ylabel('Valor MÃ­nimo')
plt.xticks(rotation=90)
plt.show()
```

```python
maximo = df_numerico.max()
print(maximo)
```

```python
maximo = df_numerico.max()
maximo.plot(kind='bar', figsize=(10, 6))
plt.title('Valor MÃ¡ximo de Variables NumÃ©ricas de Salud PÃºblica')
plt.ylabel('Valor MÃ¡ximo')
plt.xticks(rotation=90)
plt.show()
```

```python
maximo_sin_per_capita = maximo.drop('Per Capita Income')
maximo_sin_per_capita.plot(kind='bar', figsize=(10, 6))
plt.title('Valor MÃ¡ximo de Variables NumÃ©ricas (sin Per Capita Income)')
plt.ylabel('Valor MÃ¡ximo')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlacion, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de CorrelaciÃ³n de Variables NumÃ©ricas de Salud PÃºblica')
plt.show()
```
